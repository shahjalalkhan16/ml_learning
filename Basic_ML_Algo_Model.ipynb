# **Classifications:**





```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris.drop('Id', axis=1, inplace = True) # 1 or true means row
```

```python
iris.head()
```

```python
train, test = train_test_split(iris, test_size=0.2)
```

```python
train.shape
```

```python
test.shape
```

```python
train.head()
```

```python
train_X = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
train_y = train.Species
```

```python
test_X = test[['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
test_y = test.Species
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(train_X, train_y)
```

```python
prediction = model.predict(test_X)
```

```python
prediction
```

```python
accuracy = accuracy_score(prediction, test_y)
```

```python
print(f'Decision Tree model: {accuracy*100}%')
```

# Train A SVM Classifier

```python
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(train_X, train_y)
```

```python
predict = model.predict(test_X)
```

```python
predict
```

```python
accuracy_svm = metrics.accuracy_score(predict, test_y)
```

```python
print(f'Decision SVM model: {accuracy_svm*100}%')
```

Train and Evaluate a KNN model

```python
model_knn = KNeighborsClassifier()
```

```python
model_knn.fit(train_X, train_y)
```

```python
predict_knn = model_knn.predict(test_X)
```

```python
predict_knn
```

```python
accuracy_knn = metrics.accuracy_score(predict_knn, test_y)
```

```python
print(f'Accuracy of KNN model: {accuracy_knn*100}%')
```
