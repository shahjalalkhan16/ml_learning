```python
import cv2
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# **RGB vs Grayscale - Image Basics**

```python
rgb_img = cv2.imread('/kaggle/input/image-input/rgb-image.jpg')
rgb_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.title("RGB Image")
plt.show()
```

```python
print('rgb shape: ', rgb_img.shape)
rgb_img
```

```python
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.title('Gray Image')
plt.show()
```

```python
print("Gray image shape: ",gray_img.shape)
gray_img
```
